```python
import pandas as pd
```

```python
companies = pd.read_csv(r'C:\Users\bbsbh\Desktop\Python\companies.csv', index_col = 'id')
```

```python
companies.head()
```

```python
def status_func(column):
    if column<60000:
        return('Low')    
    elif column > 60000 & column <= 100000:
        return('Medium')
    else:
        return('High')
        
companies['status'] = companies['employees'].apply(status_func)
```

```python
companies.head()
```

```python
securities = pd.read_csv('securities.csv')
```

```python
securities_companies = companies.join(securities)
securities_companies.head()
```
